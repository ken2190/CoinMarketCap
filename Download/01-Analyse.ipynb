# Main

```python
for i in range(1, 11_200):
    df = pd.DataFrame(
        columns=["Price", "Volume", "FullyDilluted", "CirculatingSupply"],
    )
    df.index = df.index.rename("Timestamp")
    try:
        file = json.load(open(f"chart/{i}.json", "r"))
        dico = file["data"]["points"]
        for loc in list(dico.keys()):
            p, v, f, _, c = dico[loc]["v"]
            df.loc[loc] = [p, v, f, c]

        df = df.sort_index()
        df.to_csv(f"csv/{i}.csv")
    except:
        print(i)
```
